        from

            select 
              name, 
              stars, 
              day,
              substr(time, 1, 2) as hour1,
              substr(time, 4, 2) as minute1,

            from
            (
                select
                    name, stars, day,
                    case
                        when time like '%-1:%' then replace(time, '-', '-0')

                            when time like '1%' then time
