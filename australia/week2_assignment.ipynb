# Inspect income distributions based on age groups
## Variable Three: Age group


```python
vic_pa = vic_p.groupby(['age_low', 'income_low'])['number'].sum()
vic_p_y15_19 = vic_pa.xs(15, level='age_low')
vic_p_y20_24 = vic_pa.xs(20, level='age_low')
vic_p_y25_34 = vic_pa.xs(25, level='age_low')
vic_p_y35_44 = vic_pa.xs(35, level='age_low')
vic_p_y45_54 = vic_pa.xs(45, level='age_low')
vic_p_y55_64 = vic_pa.xs(55, level='age_low')
vic_p_y65_74 = vic_pa.xs(65, level='age_low')
vic_p_y75_84 = vic_pa.xs(75, level='age_low')
vic_p_y85_99 = vic_pa.xs(85, level='age_low')

vic_p_age_income_num = pd.concat([vic_p_y15_19, vic_p_y20_24, vic_p_y25_34, vic_p_y35_44, vic_p_y45_54, 
                                  vic_p_y55_64, vic_p_y65_74, vic_p_y75_84, vic_p_y85_99], 
                                  axis=1, 
                                  keys=['15-19', '20-24', '25-34', '35-44', '45-54', 
                                        '55-64', '65-74', '75-84', '85 over'])


vic_p_age_income_percent = vic_p_age_income_num / vic_p_pop * 100
```
